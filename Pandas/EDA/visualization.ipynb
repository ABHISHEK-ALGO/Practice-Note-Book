## 2-d plot (special case of scatter plot)

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='D',label='kohli')#solid ,dotted
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle= 'dashdot',linewidth=1,marker='*',label='sharma')
plt.title('comparison between virat and rohit sharma')
plt.xlabel('years')
plt.ylabel('batsman')
plt.legend(loc='upper right')#best in loc 
plt.ylim(0.200) #during outlier
plt.grid()
plt.show()# if graph are not plotted well in text editor since it is console its not needed

```

```python
batter = pd.read_csv("batter[1].csv")
batter
```

```python
sample = batter.head(100).sample(24,random_state=5)
#sample
```

```python
plt.figure(figsize=(15,10))
plt.scatter(sample['strike_rate'],sample['avg'],sample['runs'])
plt.xlabel('strike rate')
plt.ylabel('avg')
plt.axhline(30,color='red')
plt.axhline(40,color='green')
plt.axvline(120,color='red')

for i in range(sample.shape[0]):
    plt.text(sample['strike_rate'].values[i],sample['avg'].values[i],sample['batter'].values[i])
    
```

# sub plots

```python
fig, ax = plt.subplots(figsize=(15,7)) # two obj are made fig and ax as variable

ax.scatter(sample['batter'],sample['strike_rate'])

ax.set_title('sample plot')
ax.set_xlabel('batter')
ax.set_ylabel('strike rate')

fig.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batter['runs'],batter['strike_rate'],color='red')
ax[1].scatter(batter['runs'],batter['avg'])

ax[0].set_xlabel('runs')
ax[0].set_ylabel('strike_rate')
ax[1].set_xlabel('runs')
ax[1].set_ylabel('avg')
```

```python
plt.figure(figsize=(10,8))

fig,ax = plt.subplots(nrows=2,ncols=2,sharex=False)

ax[0,0].scatter(batter['runs'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['runs'],batter['avg'],color='green')
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])

plt.show()
```

```python
fig= plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.scatter(batter['runs'],batter['strike_rate'],color='red')

ax2 = fig.add_subplot(1,2,2)
ax2.hist(batter['avg'])
```

```python
fig = plt.figure(figsize=(10,8))

ax1 = plt.subplot(projection='3d')
ax1.scatter3D(batter['runs'],batter['avg'],batter['strike_rate'])

ax1.set_title("ipl batsman analysis")
ax1.set_xlabel('runs')
ax1.set_ylabel('avg')
ax1.set_zlabel('strike_rate')
```

```python
x = [0,2,4,6]
y = [2,4,5,7]
z = [2,5,8,4]

fig = plt.figure(figsize=(10,8))
ax = plt.subplot(projection = '3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
xx.shape

z = xx**2 + yy**2
z.shape

fig = plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
z = xx**2 + yy**2
z.shape

fig = plt.figure(figsize=(10,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
plt.colorbar(p)
```

```python
delivery = pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
# making our data set where batsman hits sixes and no. of balls not greater than 6. we are plotting heat to show on which ball most sixes are hit
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
temp_df
```

```python
grid = temp_df.pivot_table(index= 'overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
plt.figure(figsize=(10,8))
p = plt.imshow(grid)
plt.colorbar(p)
```
