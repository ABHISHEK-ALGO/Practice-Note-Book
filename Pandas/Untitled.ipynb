```python
import pandas as pd 
import numpy as np 
```

```python
emp = pd.read_csv("employee_salaries.csv")
emp.head()
```

```python
#Find the employee(s) with the second highest salary in the company

sec_high_sal = emp['salary'].drop_duplicates().nlargest(2).iloc[-1]
sec_high_sal_emp = emp[emp['salary'] == sec_high_sal]
sec_high_sal_emp
```

```python
#Find employees who earn more than the average salary of their department
avg_sal = emp.groupby('job_title')['salary'].transform('mean')
abv_avg_sal = emp[emp['salary'] > avg_sal]
abv_avg_sal
emp.shape
```

```python
# Determine the top N percent of employees based on salary
top_n_per = 0.1
sal_thresh = emp['salary'].quantile(1-top_n_per)
emp_top_10sal = emp[emp['salary'] >= sal_thresh]
lenemp_top_10sal
```
