```python
import pandas as pd 
import numpy as np 
```

```python
emp = pd.read_csv("HR_EMP_TABLE.csv")
```

```python
#Find the employee(s) with the THIRD highest EXP in the company

high_exp = emp['EXP'].drop_duplicates().nlargest(3).iloc[-1]
high_exp
high_exp = emp[emp['EXP'] == high_exp]
high_exp
```

```python
#Find employees who earn more than the average salary of their department
avg_EXP = emp.groupby('DEPT')['EXP'].transform('mean')
abv_avg_EXP = emp[emp['EXP'] > avg_EXP]
abv_avg_EXP
```

```python
# Determine the top N percent of employees based on salary
top_n_per = 0.1
exp_thresh = emp['EXP'].quantile(1-top_n_per)
emp_top_10_exp = emp[emp['EXP'] >= exp_thresh]
emp_top_10_exp

exp_thresh = emp['EXP'].quantile(1-top_n_per)
emp_top_10_exp = emp
```

```python
# Find the employee with the highest salary in each department
high_exp_per_dept = emp.iloc[emp.groupby('DEPT')['EXP'].idxmax()]
high_exp_per_dept
```

```python
avg_exp_detp = emp.groupby('DEPT')['EXP'].mean()
avg_exp_detp
```

```python
# totl emp by each dept
total_employees_per_dept = emp.groupby('DEPT').size()
total_employees_per_dept
```
