```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

df_sale = pd.read_csv("sales_datasets.csv")
df_sale.head()
df_cus = pd.read_csv("customer_datasets.csv")
df_cus.head()
```

### 1. find the name , location of customer having highest price.

```python
merge_df = pd.merge(df_cus,df_sale,on='c_id',how= 'inner',copy = False)
merge_df.head()
result = merge_df.loc[merge_df.groupby('c_id')['price'].idxmax(),['c_name_x','price','c_location']]
print(result)
```

### calculate the total transaction amount for each customer forthe current year. The output should contain Customer_Name and the total amount.

```python
df_sal = pd.read_csv("Sample - Superstore.csv")
df_sal.head(2)
df_sal.columns
```

```python
df_sal['Order Date'] = pd.to_datetime(df_sal['Order Date'],format = '%d-%m-%Y')
current_year = datetime.datetime.now().year
target_year = current_year - 10
filter_df_sal = df_sal[df_sal['Order Date'].dt.year==target_year]
result = filter_df_sal.groupby('Customer Name')['Sales'].sum().reset_index()
print(current_year)
```

```python
import mysql.connector
import pandas as pd

# Database connection configuration
config = {
    'user': 'root',
    'password': '0000',
    'host': 'localhost',
    'database': 'MySQL'
}

# Connect to the MySQL database
connection = mysql.connector.connect(**config)

# Write your MySQL query
query = """
SELECT del_partner, 
       COUNT(*) AS DELAY_ORDERS
FROM swiggy_orders
WHERE TIMESTAMPDIFF(MINUTE, ORDER_TIME, DELIVER_TIME) > PREDICTED_TIME
GROUP BY del_partner
ORDER BY DELAY_ORDERS DESC;
"""

# Execute the query and load results into a pandas DataFrame
try:
    # Execute the query and fetch results
    df = pd.read_sql(query, con=connection)
    
    # Display the dataset
    print(df)
finally:
    # Close the connection
    connection.close()
```
