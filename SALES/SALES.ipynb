```python
# FIND THE MOSST RECENT ORDER FOR EACH PRODUCT 
df_sales = pd.read_csv("Sample - Superstore.csv")
df_recent = df_sales.groupby("Product Name")["Order Date"].max().reset_index(name = "Most Recent Orders")
df_recent.sort_values(by = 'Most Recent Orders',ascending = False)
```

```python
df_sales['Year'] = pd.to_datetime(df_sales['Order Date'],format='%d-%m-%Y')
df_sales['Year'] = df_sales['Year'].dt.to_period(freq = 'Y') # df_sales['Year'] = df_sales['Year'].dt.year
yearly_sales = df_sales.groupby('Year')['Sales'].sum()
yearly_sales_growth = ((yearly_sales - yearly_sales.shift())/yearly_sales.shift())*100
yearly_sales_growth_df = yearly_sales_growth.reset_index()
yearly_sales_growth_df.columns = ['Year','Sales Growth']
yearly_sales_growth_df
```

```python
# CALCULATE THE MOVING AVERAGE FOR 3 MONTHS
df_sales['Moving Average'] = df_sales['Sales'].rolling(3).mean()
df_sales.head(4)
```

```python
# calculate difference 
df_sales['difference'] = df_sales['Sales'].diff()
df_sales.head(3)
```
