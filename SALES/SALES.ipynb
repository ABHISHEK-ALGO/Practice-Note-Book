```python
#  Calculate Month-over-Month (MoM) Sales Percentage Change
df_sal.head()
df_sal['Order Date'] = pd.to_datetime(df_sal['Order Date'],format = '%d-%m-%Y')
df_sal.info()
```

```python
df_sal['Month'] = df_sal['Order Date'].dt.to_period(freq ='M')
df_sal.head(2)
```

```python
monthly_sales = df_sal.groupby('Month')['Sales'].sum()
monthly_sales.head()
mon_change = monthly_sales.pct_change()*100
mon_change.head

# df_sal['pre_sal']= df_sal['Sales'].shift(1)
# df_sal['p_change'] = ((df_sal['Sales']-df_sal['pre_sal'])/df_sal['pre_sal'])*100
# print(df_sal)
```
